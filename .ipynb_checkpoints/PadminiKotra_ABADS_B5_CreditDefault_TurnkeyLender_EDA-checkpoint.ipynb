## Understanding Credit Card System:
1. Every month, a customer is billed as per their transactions.
2. In the next month, by the due date, an amount between the range of minimum amount due and the full amount billed is paid by the customer.
3. The next bill will be the remaining amount from previous month plus the usage in the current month.
4. Missing minimum due payments lead to late payment charges and continued misses result in default.
5. A customer is considered a defaulter when he misses payment by the due date mentioned.

```python
df['DEF_AMT'] = df['DEF_AMT'].astype(str)
plt.figure(figsize=(15,10))
plt.title("Bin wise percentages of defaulters and non-defaulters")
plt.xticks([])
plt.yticks([])
for i,col in enumerate(bill_amt_bins):    
    plt.subplot(6,1,i + 1)
    ax = sns.countplot(data = df, x = col, hue="DEF_AMT",width=0.9, palette = 'viridis')
    plt.ylim(0,13000)
    plt.ylabel('')
    plt.yticks([])
    plt.xticks([0,1,2,3,4,5,6,7,8],['<0', '= 0', '0-10k', '10k-25k', '25k-50k', '50k-75k', '75k-100k', '100k-200k', '>200k'], fontsize = 11)
    plt.xlabel(col[-7:-4])
    plt.tight_layout()
    
    defaulter_heights = [p.get_height() for j, p in enumerate(ax.patches) if j < 9]
    non_defaulter_heights = [p.get_height() for j, p in enumerate(ax.patches) if j >= 9]
    for j, p in enumerate(ax.patches):
        if j < 9:
            percentage = 100 * p.get_height() / (p.get_height()+non_defaulter_heights[j])
        else:
            percentage = 100 * p.get_height() / (p.get_height()+defaulter_heights[j-9])
        x = p.get_x() + p.get_width() / 2 - 0.15
        y = p.get_height() + 700
        ax.annotate(f"{percentage:.1f}%", (x, y), fontsize=11)
 
plt.show()
```

Observations:
- The percentages of defaults in all the bins are similar to each other. Though the number of defaulters are changing, the ratio of the amount of defaulters to the non-defaulters in within a bin are fairly constant

```python
df_diff_payment = df.copy()
df_diff_payment.head(2)
```

```python
# Subtract the previous month's payment from the bill amount for each month

df_diff_payment['AMT_DIFF_AUG'] = df['AMTBILL_AUG'] - df['PRE_SEP']
df_diff_payment['AMT_DIFF_JUL'] = df['AMTBILL_JUL'] - df['PRE_AUG']
df_diff_payment['AMT_DIFF_JUN'] = df['AMTBILL_JUN'] - df['PRE_JUL']
df_diff_payment['AMT_DIFF_MAY'] = df['AMTBILL_MAY'] - df['PRE_JUN']
df_diff_payment['AMT_DIFF_APR'] = df['AMTBILL_APR'] - df['PRE_MAY']

df_diff_payment.head(2)
```

Creating bins of the differences to understand their impact on the defaulters:

```python
diff_payments_bins=['AMT_DIFF_AUG_bin', 'AMT_DIFF_JUL_bin', 'AMT_DIFF_JUN_bin', 'AMT_DIFF_MAY_bin', 'AMT_DIFF_APR_bin']
for i,col_name in enumerate(diff_payments_bins):
    df_diff_payment[col_name]=pd.cut(df_diff_payment[diff_payments[i]],[-1500000,-1,0,10000,25000,50000,75000,100000,200000,1000000])
    df_diff_payment[col_name]=df_diff_payment[col_name].astype(str)
df_diff_payment.head()
```

```python
df_diff_payment['DEF_AMT']=df_diff_payment['DEF_AMT'].astype(int)
plt.figure(figsize=(10,10))
plt.title("Difference Bin wise distribution of defaulters")
plt.xticks([])
plt.yticks([])
for i,col in enumerate(diff_payments_bins):    
    plt.subplot(3,2,i + 1)
    ax = sns.barplot(x = col, y='DEF_AMT', data=df_diff_payment, palette = 'viridis',errorbar=None)
    plt.xticks([0,1,2,3,4,5,6,7,8],['<0', '= 0', '0-10k', '10k-25k', '25k-50k', '50k-75k', '75k-100k', '100k-200k', '>1mil'], fontsize = 8)
    plt.yticks([])
    plt.ylim(0,0.35)
    plt.xlabel(col[-7:-4])
    plt.ylabel('% of defaulters')
    plt.tight_layout()
    for p in ax.patches:
        ax.annotate("%.2f" %(p.get_height()), (p.get_x()+0.1, p.get_height()+0.005),fontsize=9)
```

##### Observations:
- The probability of defaulting is the least for those who have differences between \\$50k-\\$75k in April, May and June.
- The probability of defaulting is the least for those who have differences between \\$25k-\\$50k in July and between \\$75k-\\$100k in August.
